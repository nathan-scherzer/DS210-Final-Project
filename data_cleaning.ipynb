```python
df = pd.read_csv('nhl_game_stats.csv', index_col = 'team')
df.head()
```

```python
df = df[df["situation"] == "all"]
```

```python
columns_to_drop = ["name", "gameId", "playerTeam", "home_or_away", "gameDate", "position", "situation"]
df = df.drop(columns=columns_to_drop)
```

```python
team_mapping = {
    'ANA': 'Ducks',
    'ARI': 'Coyotes',
    'BOS': 'Bruins',
    'BUF': 'Sabres',
    'CAR': 'Hurricanes',
    'CBJ': 'Blue Jackets',
    'CGY': 'Flames',
    'CHI': 'Blackhawks',
    'COL': 'Avalanche',
    'DAL': 'Stars',
    'DET': 'Red Wings',
    'EDM': 'Oilers',
    'FLA': 'Panthers',
    'L.A': 'Kings',
    'MIN': 'Wild',
    'MTL': 'Canadiens',
    'NAS': 'Predators',
    'N.J': 'Devils',
    'NYI': 'Islanders',
    'NYR': 'Rangers',
    'OTT': 'Senators',
    'PHI': 'Flyers',
    'PIT': 'Penguins',
    'SEA': 'Kraken',
    'SJS': 'Sharks',
    'STL': 'Blues',
    'T.B': 'Lightning',
    'TOR': 'Maple Leafs',
    'VAN': 'Canucks',
    'VGK': 'Golden Knights',
    'WPG': 'Jets',
    'ATL': 'Jets',
}

df.index = df.index.to_series().replace(team_mapping)
df["opposingTeam"] = df["opposingTeam"].replace(team_mapping)
```

```python
columns_to_keep = ["team", "season", "opposingTeam",
                   "shotsOnGoalFor", "shotsOnGoalAgainst", "goalsFor", "playContinuedInZoneFor", "playContinuedOutsideZoneFor", "xGoalsFor", "highDangerShotsFor",
                   "takeawaysFor", "dZoneGiveawaysFor", "missedShotsAgainst", "goalsAgainst", "playContinuedInZoneAgainst", "xGoalsAgainst",
                   "hitsFor", "penaltiesFor", "penaltiesAgainst", "dZoneGiveawaysAgainst",
                   "faceOffsWonFor", "faceOffsWonAgainst", "playContinuedOutsideZoneAgainst", "giveawaysFor", "takeawaysAgainst", "giveawaysAgainst"]
df = df.reset_index()
df = df[columns_to_keep]
```

```python
df["penalties_for_perc"] = (df["penaltiesFor"] / (df["penaltiesFor"] + df["penaltiesAgainst"])).round(4)
df["faceoff_win_perc"] = (df["faceOffsWonFor"] / (df["faceOffsWonFor"] + df["faceOffsWonAgainst"])).round(4)

df = df.drop(columns=["penaltiesAgainst", "penaltiesFor", "faceOffsWonFor", "faceOffsWonAgainst"])

pd.set_option('display.max_columns', None)
df.head()
```

```python
df = df.rename(columns={
    "opposingTeam": "opponent",
    "shotsOnGoalFor": "shots_on_goal",
    "goalsFor": "goals",
    "playContinuedInZoneFor": "cont_in_zone_after_shot",
    "playContinuedOutsideZoneFor": "exit_zone_after_shot",
    "xGoalsFor": "exp_goals",
    "highDangerShotsFor": "dangerous_shots",
    "takeawaysFor": "takeaways",
    "shotsOnGoalAgainst": "opp_shots_on_goal",
    "dZoneGiveawaysFor": "d_zone_giveaways",
    "missedShotsAgainst": "opp_missed_shots",
    "goalsAgainst": "opp_goals",
    "xGoalsAgainst": "opp_exp_goals",
    "highDangerShotsAgainst": "opp_dangerous_shots",
    "hitsFor": "hits",
    "blockedShotAttemptsFor": "blocks",
    "giveawaysAgainst": "opp_giveaways",
    "dZoneGiveawaysAgainst": "opp_d_zone_giveaways",
    "playContinuedInZoneAgainst": "opp_cont_in_zone_after_shot",
    "playContinuedOutsideZoneAgainst": "opp_exit_zone_after_shot",
    "giveawaysFor": "giveaways",
    "takeawaysAgainst": "opp_takeaways",
})

pd.set_option('display.max_columns', None)
df.head(10)
```

```python
df.to_csv('all_game_stats.csv', index=False)
files.download('all_game_stats.csv')
```
